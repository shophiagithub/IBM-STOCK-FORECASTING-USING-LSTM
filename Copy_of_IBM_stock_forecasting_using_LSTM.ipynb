# **DATA PREPROCESSING**

```python
df['Date']=pd.to_datetime(df['Date'])
df=df.sort_values('Date')
```

```python
data=df[['Date','Open','Close','High','Low','Volume']]
```

# **DATA NORMALIZATION**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
df_scaled = pd.DataFrame(scaler.fit_transform(data.select_dtypes(include=[np.number])), columns=data.select_dtypes(include=[np.number]).columns)

fig = go.Figure()
box_color = 'turquoise'
for column in df_scaled.columns:
    fig.add_trace(go.Box(
        y=df_scaled[column].dropna(),
        name=column,
        marker=dict(color=box_color)
    ))

fig.update_layout(
    title='Box Plots of All Features',
    yaxis_title='Values',
    xaxis_title='Features',
    template='plotly_dark'
)

fig.show()
```

# **DATA SPLITTING**

```python
df_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data=train_test_split(df_scaled,test_size=0.2,shuffle=False)
```

```python
train_df=pd.DataFrame(train_data,columns=['Open','Close','High','Low','Volume'])
test_df=pd.DataFrame(test_data,columns=['Open','Close','High','Low','Volume'])
```

# **SEQUENCE GENERATION**

```python
#creating a function to generate sequences and labels with a time step of 50
def generate_sequence(df,seq_length=50):
  x=df[['Open','Close','High','Low','Volume']]
  y=df[['Open','Close','High','Low','Volume']]
  sequences=[]
  labels=[]
  for index in range(len(df)-seq_length):
    sequences.append(x.iloc[index:index+seq_length].values)
    labels.append(y.iloc[index+seq_length].values)

  sequences=np.array(sequences)  #used to convert the list of numpy arrays into a single numpy array
  labels=np.array(labels)
  return sequences,labels
```

```python
train_sequences,train_labels=generate_sequence(train_df)
test_sequences,test_labels=generate_sequence(test_df)
```

# **MODEL ARCHITECTURE**

1. Create a Sequential model.

2. Add the first LSTM layer with 50 units and return sequences. The input shape is set to (50, 5), where 50 is the sequence length and 5 is the number of features ('open', 'high', 'low', 'volume', 'close').

**LSTM Layer 1:**

Units: 50

Input Shape: (50, 5)

Return Sequences: True
3. Apply dropout regularization with a rate of 0.2 to mitigate overfitting.

**Dropout Layer 1:**

Rate: 0.2
4. Add the second LSTM layer with 50 units and return sequences.

**LSTM Layer 2:**

Units: 50

Return Sequences: True
5. Apply dropout regularization with a rate of 0.2.

**Dropout Layer 2:**

Rate: 0.2
6. Add the third LSTM layer with 50 units.

**LSTM Layer 3:**

Units: 50
7. Apply dropout regularization with a rate of 0.2.

**Dropout Layer 3:**

Rate: 0.2
8. Add a fully connected Dense layer with 5 units as the output layer.

**Dense Layer:**
Units: 5 (output)
This model architecture comprises multiple LSTM layers with dropout regularization to prevent overfitting.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dropout,Dense
```

```python
model=Sequential([
    LSTM(units=50,return_sequences=True,input_shape=(50,5)),
    Dropout(0.2),
    LSTM(units=50,return_sequences=True),
    Dropout(0.2),
    LSTM(units=50),
    Dropout(0.2),
    Dense(units=5)
])
```

# **COMPILE AND SUMMARY**

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics='mean_absolute_error')
model.summary()
```

# **MODEL TRAINING**

```python
from tensorflow.keras.callbacks import EarlyStopping
epochs=30
batch_size=32
early_stop=EarlyStopping(monitor='loss',patience=5)
history=model.fit(
    train_sequences,
    train_labels,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=(test_sequences,test_labels),
    verbose=1,
    callbacks=[early_stop]
)
```

```python
train_predictions=model.predict(train_sequences)
test_predictions=model.predict(test_sequences)
```

```python
print(history.history.keys())
```

```python
history_df = pd.DataFrame(history.history)
fig = go.Figure()

for column in history_df.columns:
    fig.add_trace(go.Scatter(x=history_df.index, y=history_df[column], mode='lines', name=column))

fig.update_layout(
    title='Loss and Mean Absolute Error over Epochs',
    xaxis_title='Epochs',
    yaxis_title='Value',
    template='plotly_dark',
    autosize=True,
    height=600,
    width=1000
)
fig.show()
```
